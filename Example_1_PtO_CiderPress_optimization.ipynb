```python
##########################
##-Imports for building-##
##########################

##-Import for ciderpress
from ciderpress.pyscf.dft import make_cider_calc

##-Imports for pyscf
from pyscf import gto, dft
from pyscf.geomopt.geometric_solver import optimize

##-General imports
import numpy as np
```

```python
####################
##-Optimizing PtO-##
####################

##-Redefining optimize function to return final energy
##-optimize() on its own does not return final energy
def optimize_with_energy(mf):
  mol_opt = optimize(mf)
  e_opt = mf.as_scanner()(mol_opt)
  return mol_opt, e_opt

##-Setting up the initial PtO molecule with def2-tzvpp basis and ecp
PtO = gto.M(
    atom = 'Pt 0 0 0; O 0 0 1.74',
    basis = 'def2-tzvpp',
    ecp = {'Pt': 'def2-tzvpp'},
    charge = 0,
    spin = 2
)

##-Creating a dft object
##-UKS for spin pol, RKS for non spin-pol.
ks = dft.UKS(PtO)

##-Making the calculator
ks = make_cider_calc(
    ks,
    ##-Choosing the CIDER functional and x, c kernels
    mlfunc = 'CIDER23X_NL_GGA.yaml',
    xkernel = 'GGA_X_PBE',
    ckernel = 'GGA_C_PBE',
    ##-Using 25% CIDER ML, 75% PBE
    ##-Does not converge with 100% ML
    xmix = 0.25
)

##-Using auxiliary basis to speed up calculations
ks = ks.density_fit()
ks.with_df.auxbasis = 'def2-universal-jfit'

##-Running the optimizer
PtO_opt, Etot_opt = optimize_with_energy(ks)

##-Saving the optimized structure as .xyz file
PtO_opt.tofile('PtO_opt.xyz', format='xyz')
```
